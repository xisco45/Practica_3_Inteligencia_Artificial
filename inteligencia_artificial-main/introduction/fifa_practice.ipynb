## Tratamientos de datos

Ya tenemos los campos que queremos analizar para el entrenamiento del modelo pero aun nos falta transformar los datos a valores que podamos analizar.

Para ello tenemos que transformar los campos que no sean numericos a numeros.

#### Funcion de tranformación de precios  
Funcion que transforma los precios, tanto valor, sueldo y clausura de restricción, 
a valores numericos


```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
   
    
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
        return ret_val
    x = x.replace('€', '')
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

#### Funcion de puntación  
Funcion que nos calcula el valor que tiene cada jugador en las diferentes posiciones que hay en el juego

```python
def puntPosicion(x):
    if type(x) == str:
        x1=x[0:2]
        x2=x[-1]
        x=int(x1)+int(x2)
        return x
    else:
        return x
```

Ahora aplicamos las funciones anteriores para las columnas necesarias

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
```

Antes de aplicar la funcion en la clausura de restriccion hay que mirar que no
haya un NAN para que no de error.  
Una NAN en la clausura de restricción indica que no tiene, lo que es equivalente a 0 por lo que cambiamos los NAN por 0

```python
df["Release Clause"].fillna((0.0),inplace=True)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
```

```python
posiciones=['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
for i in posiciones:
    df[i]=df[i].apply(puntPosicion)
    df[i].fillna((0.0),inplace=True)
df
```

Para convertir el campo club hay que hacer un tratamiento diferente al anterior.  
Este tratamiento consiste en eliminar la columna club y anadir como columnas toods los club que haya y poner 0 o 1 en en esa columna si el jugador pertenece o no a ese club

```python
clb = df.pop("Club")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)
```

Despues de tratar todos los datos puede que nos haya quedado algun NAN en algun valor. Para que no falle el entrenamiento por ese valor lo que hacemos es eliminarlo de por si.

```python
df=df.dropna()
df
```

## Prediccion

Para la prediccion final quitaremos nuestra columna de calor para compararla con el valor que nos de nuestro modelo

```python
val = df.pop("Value")
to_drop = [c for c in df.columns if not c.startswith('clb')]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=42)
```
